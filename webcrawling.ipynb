```
sections=soup.select(".xn-section")
for section in sections:
    week=section.select_one("div.xnslh-section-title").text
    titles=section.select("span.xnci-component-title")
    types=section.select("span.xnci-description-component-type")
    dic = {title.text:ftype.text for title, ftype in zip(titles, types)}
    print(week)
    for title, ftype in dic.items(): 
        print(f"{title}[{ftype}]")
    print("\n")
```
